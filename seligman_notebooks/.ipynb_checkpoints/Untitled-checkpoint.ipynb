```python
%matplotlib inline

import matplotlib
from numpy import *
from matplotlib.pyplot import *
from scipy.special import erf
#from mpl_toolkits.basemap import Basemap
#import mpl_toolkits.basemap
from mpl_toolkits.basemap import Basemap
import astropy
from math import *
from astropy.coordinates import SkyCoord
import astropy.units as unit
def ga2equ(ga):
    """
    Convert Galactic to Equatorial coordinates (J2000.0)
    (use at own risk)
    
    Input: [l,b] in decimal degrees
    Returns: [ra,dec] in decimal degrees
    
    Source: 
    - Book: "Practical astronomy with your calculator" (Peter Duffett-Smith)
    - Wikipedia "Galactic coordinates"
    
    Tests (examples given on the Wikipedia page):
    >>> ga2equ([0.0, 0.0]).round(3)
    array([ 266.405,  -28.936])
    >>> ga2equ([359.9443056, -0.0461944444]).round(3)
    array([ 266.417,  -29.008])
    """
    l = radians(ga[0])
    b = radians(ga[1])

    # North galactic pole (J2000) -- according to Wikipedia
    pole_ra = radians(192.859508)
    pole_dec = radians(27.128336)
    posangle = radians(122.932-90.0)
    
    # North galactic pole (B1950)
    #pole_ra = radians(192.25)
    #pole_dec = radians(27.4)
    #posangle = radians(123.0-90.0)
    
    ra = atan2( (cos(b)*cos(l-posangle)), (sin(b)*cos(pole_dec) - cos(b)*sin(pole_dec)*sin(l-posangle)) ) + pole_ra
    dec = asin( cos(b)*cos(pole_dec)*sin(l-posangle) + sin(b)*sin(pole_dec) )
    
    return np.array([degrees(ra), degrees(dec)])
```

```python
def galacticV_to_ra_dec(vel):
    #vel  ---  (u,v,w) vector with the U V W velocity of incoming object
    #WRT to the heliocentric system
    #function that converts a galactic velocity of an incoming object
    #to a l and b angle
    # galactic latitude and longitude
    # l - - - angle from the galactic center, l=0 points to galactic center
    # b is polar angle
    #returns  (l,b)
    (u,v,w)=vel
    u=-u
    v=-v
    w=-w
    #l = arctan(v/u)*360./(2.*pi)
    #b = arctan(w/sqrt(u**2+v**2))*360./(2.*pi)
    l=degrees(arctan2(v,u))
    b =degrees(arctan2(w,sqrt(u**2+v**2)))
    ra,dec=ga2equ((l,b))
    #ra,dec=gal_to_equat(l,b)
    return(ra,dec)
    #c = SkyCoord(l=l*unit.degree, b=b*unit.degree, frame='galactic')
   # return(c.icrs.ra.degree,c.icrs.dec.degree)

#l,b = galacticV_to_lb((u,v,w))
#print l,b
#c = SkyCoord(l=l*unit.degree, b=b*unit.degree, frame='galactic')
#print c
#print 
(u,v,w) = -11.45,-22.395,-7.746
ra, dec = galacticV_to_ra_dec((u,v,w))
print ra ,dec
#print ra.degree
```

```python
def lon2str(deg):
    min = 60 * (deg - np.floor(deg))
    deg = np.floor(deg)
    dir = 'E'
    if deg < 0:
        if min != 0.0:
            deg += 1.0
            min -= 60.0
        dir = 'W'
    return (str(int(deg/15.))+"h")
   # return ("%d\N{DEGREE SIGN} %g' %s") % (np.abs(deg),np.abs(min),dir)
```

```python
a = 1.1
b=.5
def get_de(a,b):
    a1 = -1.
    a2 = 3.
    a3 = a*b
    a4 = 2.*a-3+a*a*b*b
    a5 = 4.*a*a*b
    a6 = 2.*a*a-2.*a+1+a*a*b*b-a*a*b
    a7 = a*a*b-a*b
    return roots([a1,a2,a3,a4,a5,a6,a7])
ans =  get_de(a,b)

def get_da(x,a,b):
    return a*(1.+x*b)/(1.-x**2)-1.
#for i in arange(len(ans)):
    #if (ans[i].imag==0.):
       # print get_da(ans[i],a,b)
        #print "mod="
        #print get_da(ans[i],a,b)**2+ans[i]**2
da = get_da(ans[5],a,b)
de = ans[5]

def min_orb(rstar,fstar):
    de  =get_de(rstar,np.cos(fstar))
    mod = 10.
    for i in arange(len(de)):
        if (de[i].imag==0.):
            da = get_da(de[i],rstar,np.cos(fstar))
            mod1 = da**2+de[i]**2
            if mod1<= mod:
                mod=mod1
                delta_a = da.real
                delta_e = de[i].real
    return(delta_e,delta_a)
rstar = 1.1
fstar = pi/3.
de,da= min_orb(rstar,fstar)
print de,da
```
