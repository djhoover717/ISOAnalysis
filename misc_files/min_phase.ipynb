```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
import os

plt.rcParams["xtick.major.size"] = 9
plt.rcParams["xtick.major.width"] = 1.5
plt.rcParams["ytick.major.size"] = 9
plt.rcParams["ytick.major.width"] = 1.5
plt.rcParams["font.size"] = 14

pi = np.pi

MAIN_PATH = 'C://Users/djhoo/Documents/ISO/'
os.chdir(MAIN_PATH)
```

```python
def m(theta, H=22.4, dBO=0.75, dOS=1.):
    dBS = dBO*np.cos(theta) + np.sqrt(dOS**2 - dBO**2*np.sin(theta)**2)
    return H + 2.5*np.log10(dBS**2*dBO**2/(p(theta)*dOS**4))

def p(theta):
    return (2/3)*((1-theta/pi)*np.cos(theta) + np.sin(theta)/pi)

x = np.linspace(0, pi, 1000)
```

```python
plt.plot(x, m(x, dBO=0.25), lw=2., color='red', label=r'$d_{BO}=0.25$')
plt.plot(x, m(x, dBO=0.35), lw=2., color='green', label=r'$d_{BO}=0.35$')
plt.plot(x, m(x, dBO=0.45), lw=2., color='blue', label=r'$d_{BO}=0.45$')

plt.hlines(22., -5, pi+5, linestyles='dashed', color='black')

plt.xlabel('Phase, ' + r'$\theta$')
plt.xlim(-0.1, pi+0.1)
plt.xticks([0, pi/6, pi/3, pi/2, 2*pi/3, 5*pi/6, pi], [r'$0^{\circ}$', r'$30^{\circ}$', r'$60^{\circ}$', \
                                        r'$90^{\circ}$', r'$120^{\circ}$', r'$150^{\circ}$', r'$180^{\circ}$'])

plt.ylabel('Apparent magnitude, $m$')
plt.ylim(20, 25)

plt.legend(bbox_to_anchor=(0., 1.45), loc='upper left')

plt.savefig('figures/ecliptic_proof.pdf', bbox_inches='tight', dpi=250)
plt.show()
```
